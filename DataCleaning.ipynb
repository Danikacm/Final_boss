```python
import pandas as pd


df = pd.read_excel(r'C:\Users\danie\OneDrive\Escritorio\bases\FinalBoss\DataBases\modified\DB-grados-limpiada-georref.xlsx')

if 'Sin Vencer' not in df.columns:
    df['Sin Vencer'] = 0.0
if 'DeudaVencida' not in df.columns:
    df['DeudaVencida'] = 0.0


for tramo in ['0-30', '31-60', '61-90', '91-120', '121+']:
    if tramo not in df.columns:
        df[tramo] = 0.0


df['DeudaVencida'] = df.apply(lambda x: x['SaldoCartera'] if x['Dias Venc.'] > 0 else 0.0, axis=1)
df['Sin Vencer'] = df.apply(lambda x: x['SaldoCartera'] if x['Dias Venc.'] <= 0 else 0.0, axis=1)


df.loc[(df['Dias Venc.'] > 0) & (df['Dias Venc.'] <= 30), '0-30'] = df['DeudaVencida']
df.loc[(df['Dias Venc.'] > 30) & (df['Dias Venc.'] <= 60), '31-60'] = df['DeudaVencida']
df.loc[(df['Dias Venc.'] > 60) & (df['Dias Venc.'] <= 90), '61-90'] = df['DeudaVencida']
df.loc[(df['Dias Venc.'] > 90) & (df['Dias Venc.'] <= 120), '91-120'] = df['DeudaVencida']
df.loc[(df['Dias Venc.'] > 120), '121+'] = df['DeudaVencida']


#df.to_excel('base_financiera_corregida.xlsx', index=False)
```
